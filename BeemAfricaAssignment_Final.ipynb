```python
import requests
```

```python
url = ("https://www.mcc-mnc.com/.")
requests.get(url)
```

```python
results = requests.get(url)
print(results.text)
```

```python
from bs4 import BeautifulSoup
doc = BeautifulSoup(results.text, "html.parser")
print(doc.prettify())
```

```python
table = doc.find('table', id='mncmccTable')
# f = open('data.json', 'wb')
# f.write(data)
json_list = []
rows = table.find_all('tr')
for row in rows:
  json_data = {}
  cols = row.find_all('td')
  headers = ['MCC', 'MNC', 'ISO', 'Country', 'Country Code', 'Network']
  x = 0
  for col in cols:
    json_data[headers[x]]= col.text.strip()
    x += 1
  json_list.append(json_data)

json_list.pop(0)
```

```python
import json
with open('data.json', 'w') as f:
    json.dump(json_list, f)
```
